<h1>Part 1: Dataset</h1>
I choose World's Spotify top 50 playlist 2023 dataset from kaggle. There are 73 countries and a global section available in this dataset. It also shows different aspect of these songs. This is meaningful for me because I use Spotify to listening music everyday so I want to know if my favorite songs are popular. I want to know which songs are popular and who are the popular artists globally. Then I want to know if they are also popular in the US. I want to know if certain features of a song influence their popularity and will these features determine their popularity in the future. 

```python
#Load the dataset
import pandas as pd
df=pd.read_csv('Top-50-musicality-global.csv')
df
```

<h1>Part 2: Exploratory Data Analysis</h1>
In this part I will find out who are the top artists by looking at the top 50 songs globally and in the US.I will look up the top 50 most popular songs globally. Then I will look for the most popular artists in each areas by comparing how many times they've appeared in the following subsets in each areas.

```python
#find top50 songs globally
top50song_global=df[df["Country"]=="Global"]
top50song_global
```

```python
popularartist_global=top50song_global[["Artist Name","Popularity"]]
agg_count_global=popularartist_global.groupby("Artist Name").agg("count").reset_index()
top10artist_global=agg_count_global.sort_values(by='Popularity')
top10artist_global
```

```python
top10song_global=top50song_global.nlargest(10,"Popularity")
top10song_global
```

```python
top50song_us=df[df["Country"]=="USA"]
top50song_us
```

```python
popularartist_us=top50song_us[["Artist Name","Popularity"]]
agg_count_us=popularartist_us.groupby("Artist Name").agg("count").reset_index()
top10artist_us=agg_count_us.sort_values(by='Popularity')
top10artist_us
```

<h1>Part 4: Data Science</h1>

In this part I will explore will certain features of a song affect it's popularity. I believe there is some connection between the popularity of a song and some characteristics of the song. I am also wondering if these features will influence the popularity of the world's 10 most popular songs in the future.First I will use corr.()function to find the correlation coefficient of numerical features of all songs. Then I will use model.predict() function to predict popularity of the world's most 10 most popular songs. 

```python
numerical_column=["Popularity","Danceability","Acousticness","duration","Energy","Instrumentalness","Key","Liveness","Loudness","Mode","Speechiness","Tempo","TSignature","Positiveness"]
numerical_column
```

```python
df_numerical=df[numerical_column]
df_numerical
```

```python
df_numerical.corr()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
ind=["Loudness","TSignature","Speechiness"]
dep="Popularity"
model.fit(top10song_global[ind],top10song_global[dep])
top10song_global["Predicted Popularity"]=model.predict(top10song_global[ind])
top10song_global
```
